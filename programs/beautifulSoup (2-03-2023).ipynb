```python
# Assignment
# Date,Particulars, amount type, Amount, Balance amount
from bs4 import BeautifulSoup
import requests
required_list=[]
with open('output15.html') as html_file:
    soup = BeautifulSoup(html_file,'lxml')
    
for article in soup.find_all('tr',class_='listwhiterow'):
    inner_list=[]
    for sl in article.find_all('td'):
        
        inner_list.append(sl.text)
    required_list.append(inner_list)

print("Date\tParticulars\tamount\ttype\tAmount\tBalance\tamount")
for i in required_list:
    print(i[1],'\t',i[3],'\t',i[5],'\t',i[6],'\t',i[7])
```
